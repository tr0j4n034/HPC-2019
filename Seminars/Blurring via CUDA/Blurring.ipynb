```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# amazing sample data
img_data = np.array([[0.5, 0.6, 0.7, 0.8],
                     [0.1, 0.2, 0.3, 0.4],
                     [0.0, 0.1, 0.2, 0.3],
                     [0.9, 0.8, 0.7, 0.6]])
```

```python
kernel = np.array([[0.04, 0.12, 0.04],
                   [0.12, 0.36, 0.12],
                   [0.04, 0.12, 0.04]])
```

```python
def blur(img, kernel, iters): # slow & furious
    for it in range(iters):
        img_out = np.zeros(img.shape)
        for i in range(img.shape[0]):
            for j in range(img.shape[1]):
                for dx in range(-1, 2):
                    for dy in range(-1, 2):
                        if i + dx < 0 or i + dx >= img.shape[0]: continue
                        if j + dy < 0 or j + dy >= img.shape[1]: continue
                        img_out[i][j] += img[i + dx][j + dy] * kernel[dx + 1][dy + 1];
        img = img_out.copy()
    return img
```

```python
print(img_data)
```

```python
print(blur(img_data, kernel, 1))
```

```python
print(blur(img_data, kernel, 2))
```
